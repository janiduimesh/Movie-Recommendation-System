```python
import requests
import os

# TMDb API endpoint for movie search
api_url = "https://api.themoviedb.org/3/search/movie"

# TMDb API key (you need to get your own key from TMDb)
api_key = "87b2b485db0dee88c42d7317ade4d92f"
```

```python
import pandas as pd

movies = pd.read_csv('movies.csv')
movie_titles = list(movies['title'])
print(len(movie_titles))

# movie_titles = movie_titles[1:20]
movie_titles
```

```python
import requests

movie_data = []

for movie_title in movie_titles:
    # Remove the year from the movie title
    movie_title = movie_title.split(' (')[0]

    # Make a request to TMDb API to search for the movie
    params = {
        "api_key": api_key,
        "query": movie_title
    }

    try:
        response = requests.get(api_url, params=params)
        
        # Check if the request was successful
        if response.status_code != 200:
            print(f"Error: Failed to fetch data for {movie_title}. Status code: {response.status_code}")
            movie_data.append({"Title": movie_title, "Poster URL": "Not found"})
            continue
        
        try:
            data = response.json()
        except ValueError as e:
            print(f"Error decoding JSON for {movie_title}: {e}")
            print(f"Response content: {response.text}")
            movie_data.append({"Title": movie_title, "Poster URL": "Not found"})
            continue
        
        # Check if 'results' is in the JSON response
        if "results" in data and data["results"]:
            # Get the poster path for the first result
            poster_path = data["results"][0].get("poster_path")
            
            if poster_path:
                # Construct the URL for the poster and add it to the list
                poster_url = f"https://image.tmdb.org/t/p/original{poster_path}"
                movie_data.append({"Title": movie_title, "Poster URL": poster_url})
            else:
                movie_data.append({"Title": movie_title, "Poster URL": "Not found"})
        else:
            # If no results or 'results' key not present, add only the title to the list
            movie_data.append({"Title": movie_title, "Poster URL": "Not found"})
    
    except requests.exceptions.RequestException as e:
        print(f"Request failed for {movie_title}: {e}")
        movie_data.append({"Title": movie_title, "Poster URL": "Not found"})

# Print the movie_data to see the results
for movie in movie_data:
    print(movie)
```

```python
movie_data_df = pd.DataFrame(movie_data)
movie_data_df = movie_data_df.rename(columns={'Title': 'title'})
movie_data_df
```

```python
# Create a new Series with the movie titles
movie_titles_series = pd.Series(movie_titles, name='title')

# Replace the "Title" column in the DataFrame with the new Series
movie_data_df['title'] = movie_titles_series
movie_data_df
```

```python
# merge dfs
merged_df_url = movies.merge(movie_data_df, how='inner', on='title')
merged_df_url
```

```python
## write to a csv file
merged_df_url.to_csv('movie_data_with_urls.csv', index=False)  # Set index=False to exclude the index column
```

### Handle not found Images

```python
no_image_list = []

for i in movie_data:
    if i['Poster URL'] == 'Not found':
        no_image_list.append(i['Title'])

no_image_list
```

```python
# Remove ', The' from each string in the list
no_image_list = [title.replace(', The', '') for title in no_image_list]
no_image_list
```

```python
movie_data_new = []

## get the image URL which are not returned by the above code
for title in no_image_list:
    # Remove the year from the movie title
    # movie_title = movie_title.split(' (')[0]

    # Make a request to TMDb API to search for the movie
    params = {
        "api_key": api_key,
        "query": title
    }
    response = requests.get(api_url, params=params)
    data = response.json()

    if data["results"]:
        # Get the poster path for the first result
        poster_path = data["results"][0]["poster_path"]
        
        # Construct the URL for the poster and add it to the list
        poster_url = f"https://image.tmdb.org/t/p/original{poster_path}"
        movie_data_new.append({"title": title, "Poster URL": poster_url})
    else:
        # If poster not found, add only the title to the list
        movie_data_new.append({"title": title, "Poster URL": "Not found"})

movie_data_new
```
